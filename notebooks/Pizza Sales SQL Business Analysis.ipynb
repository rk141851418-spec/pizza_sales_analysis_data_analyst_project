##### In this notebook, SQL is used to analyze the final analytical dataset and answer key business questions related to pizza sales performance. The objective is to uncover meaningful insights into customer purchasing behavior, product performance, and revenue trends to support data-driven decision-making.


###  Importing Libraries

```python
import pandas as pd
from sqlalchemy import create_engine
import warnings
warnings.filterwarnings('ignore')
```

### Creating Database Connection

```python
engine = create_engine('sqlite:///pizza_sales_analysis.db')
df = pd.read_sql_query('select * from pizza',engine)
```

### Business Questions

#### Overall Performance
##### What is the total number of orders, pizzas sold, and total revenue?

```python
pd.read_sql_query('select count(distinct order_id) as orders ,sum(quantity) as quantity,sum(revenue) as revenue from pizza',engine)
```

##### The business recorded 21,350 total orders, resulting in the sale of 49,574 pizzas and generating a total revenue of $817,860.05. This indicates strong overall sales performance and provides a solid foundation for deeper analysis of product demand, customer behavior, and revenue trends.

#### Product Performance

##### Which pizza sells the most by quantity?

```python
pd.read_sql_query('select name,sum(quantity) as quantity from pizza group by name order by quantity desc  limit 10',engine)
```

##### Which pizza sells the least?

```python
pd.read_sql_query('select name,sum(quantity) as most_sold  from pizza group by name order by most_sold limit 1 ',engine)
```

##### Which pizza generates the most revenue?

```python
pd.read_sql_query('select name,sum(revenue) as revenue from pizza group by name order by revenue desc  limit 10',engine)
```

##### The Classic Deluxe Pizza is the most sold pizza by quantity.
##### The Thai Chicken Pizza is the highest revenue-generating pizza.
##### The top-selling pizzas by quantity and revenue are not exactly the same, showing the impact of pricing and product mix.
##### Sales are well distributed across multiple pizzas, indicating balanced customer demand rather than reliance on a single product.

#### Size Analysis

##### How does revenue vary by pizza size?

```python
pd.read_sql_query('select size,sum(revenue) as revenue from pizza group by size order by revenue desc ',engine)
```

##### What is the average order quantity by pizza size?

```python
pd.read_sql_query('select size,avg(quantity) as avg_quantity from pizza group by size ',engine)
```

##### Large (L) pizzas generate the highest revenue ($375,318.70), contributing a significant share of total sales.
##### Medium and Small sizes follow, while XL and XXL contribute very little revenue, indicating low demand for extra-large sizes.
##### The average quantity per order is close to 1 for all sizes, showing customers typically buy one pizza per order item.

#### Category Analysis

##### Which pizza category sells the most by quantity?

```python
pd.read_sql_query('select category,sum(quantity) as salles  from pizza group by category order by salles desc ',engine)
```

##### Which pizza category generates the highest revenue?

```python
pd.read_sql_query('select category,sum(revenue) as rev  from pizza group by category order by rev desc ',engine)
```

##### Classic category sells the most pizzas (14,888 units) and also generates the highest revenue ($220,053.10).
##### Supreme, Chicken, and Veggie categories perform similarly in revenue, showing balanced demand across categories.

#### Time-Based Analysis

##### How does revenue vary by hour of the day?

```python
pd.read_sql_query('select hour,sum(revenue) as revenue  from pizza group by hour order by revenue desc ',engine)
```

##### Sales peak during lunch (12–1 PM) and evening (5–7 PM) hours.
##### Revenue drops significantly after 10 PM, with minimal late-night sales.
##### This clearly highlights lunch and dinner as the busiest business periods.

##### How does revenue vary by month?

```python
pd.read_sql_query('select month_name,sum(revenue) as revenue  from pizza group by month_name order by revenue desc ',engine)
```

##### Revenue is fairly consistent across months, showing stable demand throughout the year.
##### July and May are the highest revenue months, while October and September show slightly lower performance.

#### Customer Behavior

##### How do weekend sales compare to weekday sales?

```python
pd.read_sql_query('select weekday, sum(revenue) as revenue  from pizza   group by weekday  order by revenue desc ',engine)
```

##### Friday generates the highest revenue, followed by Thursday and Saturday.
##### Sunday has the lowest revenue, suggesting weaker demand at the end of the weekend.
##### Overall, sales are stronger toward the end of the workweek.
